```python
import numpy as np
import pandas as pd
from google.colab import files 
import io
```

```python
file=files.upload()
```

```python
data=pd.read_csv(io.StringIO(file['realest.csv'].decode('utf-8')))
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=['Price','Bedroom','Space','Room','Lot','Tax','Bathroom','Garage','Condition'],inplace=True)
data.head()
```

```python
data.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = ['Bedroom','Space','Room','Lot','Tax','Bathroom','Garage','Condition']
X = data[linear]
y = data['Price']
lm = LinearRegression()
model = lm.fit(X, y)
```

```python
model.score(X,y)
```

```python
model.coef_
```
